```python
news_headline=[]
for headline in table9.find_all("a",attrs={"class": "LatestNews-headline"}):
    print(headline.text)
    news_headline.append(headline.text)
    #print(news_time)
```

```python
news_link=[]
for link in table9.find_all('a', 
                          attrs={'href': re.compile("^https://")}):
    # display the actual urls
    print(link.get('href')) 
    news_link.append(link.get('href'))
```
